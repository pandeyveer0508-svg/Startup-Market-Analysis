```python
import matplotlib.pyplot as plt 
import seaborn as sb
from sklearn.preprocessing import MinMaxScaler
sk=MinMaxScaler()
data=sk.fit_transform(startup_data[["Revenue","Product_Uniqueness_Score"]])
product_scaled_dataset=pd.DataFrame(data,columns=["New_Revenue_Scaled","New_Product_Uniqueness_Score_Scaled"])
plt.figure(figsize=(12,6))
sb.scatterplot(x=product_scaled_dataset["New_Revenue_Scaled"],y=product_scaled_dataset["New_Product_Uniqueness_Score_Scaled"],data=product_scaled_dataset)
plt.xlabel("Product_Uniqueness_Score")
plt.ylabel("Company_Revenue")
plt.title("Relation Between Product_uniqueness_score And Revenue")
plt.show()
```

```python
industry_revenue_record=startup_data.groupby("Industry")["Revenue"].sum().sort_values(ascending=False)
industry_revenue_record
```
